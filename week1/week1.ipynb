2D arrays

1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

3. Determine the minimum; maximum; average; and standard deviation of all entries.

4. Sort the DataFrame: by rows, by columns, on row 2, on column 2.

*Note â€“ you should first write your program with a smaller array (perhaps [3,2] to make sure you get it running correctly. Then modify your program to meet the size requirements [10,50].

```python
# 1. Create an array of size [10,50] and 
# use the random function to fill it with integers between 1 and 500 (inclusive).
import numpy as np
import pandas as pd

rows = 10
cols = 50
minVal = 1
maxVal = 501 # excludes this

arr = np.random.randint(minVal,maxVal,rows*cols)

print(arr)

newarr = arr.reshape(rows, cols)

print(newarr)

# Load it to a DataFrame.

index = ['Row'+str(i) for i in range(1, rows+1)]
columns = ['Column'+str(i) for i in range(1, cols+1)]

#print(index)
#print(columns)

df = pd.DataFrame(data=newarr, index=index, columns=columns)
#df = pd.DataFrame.from_records(arr)

print(df)

```

```python
# 2. Calculate the sum of each row; the sum of each column; and the sum of all entries.
print("Sum")
print(df.sum(axis=1))
print(df.sum(axis=0))
print("Sum of all entries:",df.sum().sum())
```

```python
# 3. Determine the minimum; maximum; average; and standard deviation of all entries.
print("Minimum of all entries:", df.min().min())
print("Maximum of all entries:", df.max().max())
# df.mean().mean() also works
print("Average of all entries:", df.stack().mean())
print("Standard Deviation of all entries:", df.stack().std())
```

```python
# 4. Sort the DataFrame: by rows, by columns, on row 2, on column 2.
print("Actual DF")
print(df)

print("Sort by Row2")
print(df.sort_values(by = 'Row2', axis = 1))

print("Sort by Column2")
print(df.sort_values('Column2'))
```
