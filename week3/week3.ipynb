### First creating a sample dataset of Students in a class
Overall Students = 60
List of Subjects = Physics, Chemistry, Maths, English, Computer Science, Social Studies
Count of Students who scored 90-100 Marks in each subject - P-5, C-4, M-8, E-11, CS-7, SS-11


### Analysis Question - How well is class performing in each of the Subjects?

```python
# importing required libraries packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creating Sample Data , based on the above statement
subjects = ['Physics','Chemistry','Maths','English', 'Computer Science','Social Studies']
studentCount = [5, 4, 8, 11, 7, 11]
```

```python
# Line Plot
plt.figure(figsize=(16,9))
plt.plot(subjects, 
        studentCount);
plt.plot(subjects, 
        studentCount,
         'o');
plt.grid(True)
plt.title("Students scoring above 90% in subject")
plt.ylabel("Count of Students")
plt.xlabel("Subjects")
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(16,9))
plt.bar(subjects, 
        studentCount);
plt.grid(True)
plt.title("Students scoring above 90% in subject")
plt.ylabel("Count of Students")
plt.xlabel("Subjects")
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(10,10))
plt.pie(studentCount, 
        labels=subjects, 
        autopct = lambda p:f'{p:.1f}%', 
        wedgeprops={"edgecolor":"black",
                    'linewidth': 3,
                    'linestyle': 'solid', 
                    'antialiased': True});
plt.title("Students scoring above 90% in subject")
```

Let's say we first try to understand when we use each one of them.

<b>When to use a Line Plot?</b>
Line plots are used to track changes over short and long periods of time or we may say changes over a fixed value which increases or is continuous in nature. They may be used to measure the progress over the same x axis value for multiple categories. So we may say we are plotting values against some continuous number line and the y values represent the frequency of some event at that value.

<b>When to use a Bar Chart?</b>
Bar graphs are used to compare things between different groups and these groups are categorical. So the group place can be changed on the chart without affecting the meaning of the visualization.

<b>When to use a Pie Chart?</b>
Pie charts are a visual way of displaying data that might otherwise be given in a small table.
Pie charts are useful for displaying data that are classified into nominal or ordinal categories.
Pie charts are generally used to show percentage or proportional data and usually the percentage represented by each category is provided next to the corresponding slice of pie.
